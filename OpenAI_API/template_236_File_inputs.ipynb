# 파일 입력
비전 기능을 갖춘 OpenAI 모델은 PDF 파일을 입력으로 사용할 수도 있습니다. 

모델이 PDF 콘텐츠를 이해할 수 있도록 각 페이지의 추출된 텍스트와 이미지를 모델의 컨텍스트에 추가합니다. 모델은 텍스트와 이미지를 모두 사용하여 응답을 생성할 수 있습니다. 이는 예를 들어 다이어그램에 텍스트에 없는 주요 정보가 포함된 경우 유용합니다.

먼저 Files API를 사용하여 PDF를 업로드한 다음, 모델에 대한 API 요청에서 해당 파일 ID를 참조합니다.

## 파일 URL (File URLs) 사용

PDF 파일을 직접 업로드하지 않고, 공개된 외부 PDF 링크(URL)를 통해 모델이 해당 파일에 접근하도록 설정할 수 있습니다.

### 파일 업로드 (Uploading files)

먼저 Files API를 사용하여 PDF 파일을 업로드한 다음, 해당 파일의 파일 ID(file ID) 를 모델의 API 요청에서 참조

### Base64로 인코딩된 파일
PDF 파일 입력도 Base64로 인코딩된 입력으로 보낼 수 있습니다.

---------------------------------
### 사용 시 고려사항 (Usage considerations)

- 토큰 사용량 (Token usage)

    모델이 PDF 내용을 이해할 수 있도록, OpenAI는 각 페이지의 추출된 텍스트와 페이지 이미지를 모두 모델의 컨텍스트에 포함시킵니다. 따라서 PDF를 입력으로 사용할 경우 토큰 사용량과 비용이 증가할 수 있습니다.

- 파일 크기 제한 (File size limitations)

    한 번에 여러 개의 파일을 업로드할 수 있습니다. 단, 각 파일은 50MB 미만이어야 합니다.  
    하나의 API 요청 내에서 전체 파일 용량의 합도 50MB를 초과할 수 없습니다.
